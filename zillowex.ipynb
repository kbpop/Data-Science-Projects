```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
cmap = plt.cm.get_cmap("viridis")


for count, i in enumerate(graph_features):

    fig, ax = plt.subplots(1, 1)
    plt.Figure(figsize=(5,5))
    divider = make_axes_locatable(ax)
    plt.xlim(-130, -65)
    plt.ylim(22, 52)
    # cax = divider.append_axes("right", size="5%", pad=0.1)
    # plt.legend(title="My Title", fontsize=5, title_fontsize=15)
    plt.title('Graph Showing the U.S. ' +  i, fontsize=40)
    # pos = ax.imshow()
    cax = divider.append_axes("right", size="4%", pad=-3)
    mapping.plot(column = i,ax=ax, legend=True, cax=cax, cmap=cmap)
    # plt.colorbar(label="Like/Dislike Ratio", orientation="horizontal")
    # adm_gdf.plot(column='pop_est', ax=ax, legend=True, cax=cax)

    plt.gcf().set_size_inches(30, 30)
    plt.savefig(str(i).replace('/',' ') +'.png', bbox_inches='tight')
    # counties.plot(ax=ax, legend=True, cax=cax)
```
